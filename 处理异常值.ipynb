```python
import numpy as np
import matplotlib.pyplot as plt
data=[0,-(0.47)/100,-6.64/100,12.18/100,-3.62/100,0.56/100,-10.83/100,0,-2.09/100]
plt.plot(data)
plt.show()
x=np.arange(9)
```

```python
data2=[1]
for i in range(8):
    temp=data2[i]*(1+data[i+1])
    data2.append(temp)
```

```python
plt.plot(data2)
plt.rcParams['font.sans-serif']=['SimHei']#解决乱码问题
plt.title('未修正前曲线')
plt.show()
```

```python
data3=data2[:]#备份
for i in range(6,9):
    data3[i]=data3[i]*(1+0.03)
```

```python
plt.plot(data3)
plt.title('修正后曲线')
plt.show()
plt.plot(x,data2,label='修正前')
plt.plot(x,data3,label='修正后')
plt.legend(loc=0)
plt.show()
```
